```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import date 
pathcsv = 'Survey Dataset - BMS_Amazon.csv'
pathxlsx = 'Survey Dataset - BMS_Amazon.xlsx'
data = pd.read_excel(pathxlsx, index_col='Sr. No.')

data = data[pd.notnull(data['Full Name'])]
data['DOB'] = pd.to_datetime(data['DOB'])

def countAge(birthDate):
    today = date.today()
    age = today.year - birthDate.year - ((today.month, today.day) < (birthDate.month, birthDate.day))
    return age
data['Age'] = data['DOB'].apply(lambda x: countAge(x))
#data['Age'] = data['Age'].astype(int)
data = data.apply(lambda x : x.astype(str).str.upper())  #All data in string in upper case
data['Age'] = pd.to_numeric(data['Age'])                 #Numerical value 
data['Work_Exp'] = pd.to_numeric(data['Work_Exp'])        #Numerical value 
data['Work_Exp'] = data['Work_Exp'].apply(lambda x : round(x))    #Numerical value round 

def sportType(strSport):
    if(strSport == 'FOOTBALL' or strSport == 'CRICKET' or strSport == 'BOTH' or strSport == 'NEITHER'):
        return strSport
    else:
        return "NEITHER"
data['Sport'] = data['Sport'].apply(lambda x : sportType(x))

data['Fav_sport_game'] = data['Fav_sport_game'].str.replace(" ","")


def chkFoodType(strTemp):
    if(strTemp == 'NON-VEG' or strTemp == 'VEG'):
        return strTemp
    elif strTemp.find('NON') != -1:
        if(strTemp.find('-') == -1):
            return('NON-VEG')
data['Food_Pref'] = data['Food_Pref'].apply(lambda x : chkFoodType(x))


data.head(10)
```

```python
vegNonVeg = data.Food_Pref.value_counts()
# label the axes
plt.figure(figsize=(14,15))
plt.xlabel("Veg or Non-Veg")
plt.ylabel("No. of User's")
plt.title("Veg Vs. Non-Veg")
# Rotate X-axes labels
plt.xticks(rotation=45)
plt.yticks(rotation=45)
vegNonVeg.plot(kind = 'bar')
# bar chart
plt.show()
```
