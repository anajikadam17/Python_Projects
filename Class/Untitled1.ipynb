### Read the data set

```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
data = pd.read_csv('Data.csv') 
data.head(37)
```

```python
data['Voucher_Pref'] = data['Voucher_Pref'].str.upper() 
data.head(15)
```

```python
len(set(data['Full Name']))
```

```python
data['Work_Exp'] = data['Work_Exp'].round()
data['DOB'] = pd.to_datetime(data['DOB'])
data['age'] = (pd.to_datetime('today')-pd.to_datetime(data['DOB']))/365.25

data['age']=data['age'].dt.days.astype('int16')
data.head(37)

```

```python
#age Distribution on histogram
from matplotlib import pyplot as plt
import numpy as np
plt.figure(figsize = (14,8))
plt.hist(data['age'])
plt.title('Distribution of Age')
plt.xlabel('age')
plt.ylabel('Counts')
plt.show()
```

```python
#age Distribution on histogram
from matplotlib import pyplot as plt
import numpy as np
plt.figure(figsize = (14,8))
plt.hist(data['Work_Exp'])
plt.title('Distribution of Work_Exp')
plt.xlabel('Work_Exp')
plt.ylabel('Counts')
plt.show()
```

```python
#Bar chart for age and food pref
re=data.groupby(['age','Food_Pref'])['Food_Pref'].count()
re.unstack().plot(kind='bar', stacked=True, figsize=(15,10))
plt.xlabel('AGE')
plt.ylabel('Food pref')
plt.xticks(rotation=45)
plt.show()
```

```python

plt.figure(figsize = (14,8))
plt.scatter(x=data['age'],y=data['Full Name'])

plt.xlabel('name')
plt.ylabel('age')
plt.xticks(rotation=90)
plt.show()
```

```python
u=data[data['Fav_sport_game']=='CRICKET']
x = data[data['Fav_sport_game'] == 'FOOTBALL']
y = data[data['Food_Pref'] == 'NEITHER']
z= data[data['Fav_sport_game'] == 'BOTH']
labels = ["CRICKET","FOOTBALL","NEITHER","BOTH"]
sizes = [len(u),len(x),len(y),len(z)]
colors = ['lightcoral', 'lightskyblue','blue','red']
explode = (0.1,0.4,0.6,0.8)
plt.title('Sports')
plt.pie(sizes, explode=explode, labels=labels, colors=colors,autopct='%1.1f%%', shadow=True)
plt.show
```
