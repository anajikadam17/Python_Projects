```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('flights')
df.head()
```

```python
sns.relplot(x = 'passengers',y = 'month', data=df) #statistical relationship data
```

```python
sns.relplot(x = 'passengers',y = 'month', hue = 'year', data=df)
```

```python
df1 = sns.load_dataset('tips')
df1.head()
```

```python
sns.relplot(x = 'time', y= 'tip', data=df1, kind='line') #categorical data
```

```python
sns.catplot(x = 'day', y= 'total_bill', data=df1)
```

```python
sns.catplot(x = 'day', y= 'total_bill', data=df1, kind='violin')
```

```python
sns.catplot(x = 'day', y= 'total_bill', data=df1, kind='boxen')
```
